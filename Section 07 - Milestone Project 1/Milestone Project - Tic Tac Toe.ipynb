```python
def display_board(game_board):
    
    row1 = game_board[:3]
    row2 = game_board[3:6]
    row3 = game_board[6:]
    
    print(" | ".join(row1))
    print(" | ".join(row2))
    print(" | ".join(row3))
```

```python
def verify_victory(game_board):
    
    possible_win = ["XXX", "OOO"]
    
    # the rows
    row1 = "".join(game_board[0] + game_board[1] + game_board[2])
    row2 = "".join(game_board[3] + game_board[4] + game_board[5])
    row3 = "".join(game_board[6] + game_board[7] + game_board[8])
    
    # the columns
    column1 = "".join(game_board[0] + game_board[3] + game_board[6])
    column2 = "".join(game_board[1] + game_board[4] + game_board[7])
    column3 = "".join(game_board[2] + game_board[5] + game_board[8])
    
    # the diagonals
    diag1 = "".join(game_board[0] + game_board[4] + game_board[8])
    diag2 = "".join(game_board[2] + game_board[4] + game_board[6])
    
    return (
        # checking rows
        row1 in possible_win or row2 in possible_win or row3 in possible_win or 
        
        # checking columns
        column1 in possible_win or column2 in possible_win or column3 in possible_win or 
        
        # checking diagonals
        diag1 in possible_win or diag2 in possible_win
    )
    
```

```python
def player_turn(player, game_board):
    
    position = input(f"Player {player}'s turn: ")
        
    spots = ["1", "2", "3", "4", "5", "6", "7", "8", "9"]
    
    while position not in spots:
        position = input("Invalid position! Please try again: ")
        
    position = int(position)
    
    while game_board[position - 1] != " ":
        position = input("Position is occupied! Please try again: ")
        
        while position not in spots:
            position = input("Invalid position! Please try again: ")
        
        position = int(position)
    
        if game_board[position - 1] != " ":
            continue
    
    if player == 1:
        game_board[position - 1] = "X"
    else:
        game_board[position - 1] = "O"
        
    return game_board
    
```

```python
def verify_tie(game_board):
    return " " not in game_board
```

```python
def restart_game():
    game_board = [" 1", " 2", " 3", " 4", " 5", " 6", " 7", " 8", " 9"]
    print("Player 1 is X, player 2 is O. Player 1 starts.")
    print("The board is as follows: ")

    display_board(game_board)
    print("-------------")
```

```python
# setting up the game
game_board = [" ", " ", " ", " ", " ", " ", " ", " ", " "]
# game_board = [" 1", " 2", " 3", " 4", " 5", " 6", " 7", " 8", " 9"]
victory = False
tie = False
player = 1

restart_game()
# print("Player 1 is X, player 2 is O. Player 1 starts.")
# print("The board is as follows: ")

# display_board(game_board)
# game_board = [" ", " ", " ", " ", " ", " ", " ", " ", " "]
# print("-------------")

# running the game
while victory == False:
    
    game_board = player_turn(player, game_board)
    
    display_board(game_board)
    
    victory = verify_victory(game_board)
    
    
    if victory:
        print(f"Player {player} wins!")
        restart = input("Play again? (Y/N): ")
        
        while restart not in ["y", "n", "Y", "N"]:
            restart = input("Error! Please try again. Play again? (Y/N): ")
        
        if restart in ["y", "Y"]:
            game_board = [" ", " ", " ", " ", " ", " ", " ", " ", " "]
            victory = False
            tie = False
            player = 1
            restart_game()
            continue
        
    else:
        tie = verify_tie(game_board)
        
        
    if tie:
        restart = input("It's a tie! Play again? (Y/N): ")
        
        while restart not in ["y", "n", "Y", "N"]:
            restart = input("Error! Please try again. Play again? (Y/N): ")
        
        if restart in ["y", "Y"]:
            game_board = [" ", " ", " ", " ", " ", " ", " ", " ", " "]
            victory = False
            tie = False
            player = 1
            restart_game()
            continue
    
    if player == 1:
        player = 2
    else:
        player = 1
    
```
